# 그리드서치, 랜덤서치, 베이지안서치로 최적 파라미터 찾기

```python
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import RepeatedStratifiedKFold
from sklearn.ensemble import RandomForestClassifier
```

```python
params = {'max_depth' : [10, 20, 30], 'n_estimators': [300, 400, 500], 'min_samples_leaf' : [1,2] }
```

## 그리드 서치
<ul>
<li>Manual Search에 비해, Grid Search와 Random Search는 체계적인 방식으로 Hyperparameter Optimization을 수행</li>
<li>Grid Search는 모든 parameter의 경우의 수에 대해 cross-validation 결과가 가장 좋은 parameter를 고르는 방법</li>
<li>전체 탐색 대상 구간, 간격은 사람의 손이 필요하나 균등하고 전역적인 탐색이 가능하다는 장점이 있음</li>
<li>단 hyperparameter의 개수가 많아질수록 전체 탐색 시간이 기하급수적으로 증가</li>
</ul>

```python
grid_cv = GridSearchCV(model, param_grid = params, cv = 3, n_jobs = 3, verbose=3)
grid_cv.fit(X_train, y_train)
pred = grid_cv.predict(X_test)
print(pred)
print(sorted(grid_cv.cv_results_.keys()))
print('최적 하이퍼 파라미터: ', grid_cv.best_params_)
print('최고 예측 정확도: {:.4f}'.format(grid_cv.best_score_))
```

## 랜덤서치
<ul>
<li>Grid Search가 전수조사라면 Random Search는 표본조사</li>
<li>hyperparameter 값을 랜덤 샘플링(sampling)해 선정</li>
<li>Random Search는 Grid Search에 비해 불필요한 반복 수행 횟수를 대폭 줄이면서, 동시에 정해진 간격(grid) 사이에 위치한 값들에 대해서도 확률적으로 탐색이 가능하므로, 최적 hyperparameter 값을 더 빨리 찾을 수 있다</li>
<li>Grid Search는 Unimportant parameter와 Important Parameter를 동일하게 관측해야하기 때문에, 정작 Important Parameter를 다양하게 시도해볼 수 있는 기회가 적지만 Random Search는 grid로 제한되지 않기 때문에 확률적으로 Important Parameter를 더 살펴볼 수 있는 기회가 있음</li>
</ul>

```python
rand_cv = RandomizedSearchCV(model, param_distributions = params, n_iter = 10, cv = 3, n_jobs = 3, verbose=3)
rand_cv.fit(X_train, y_train)
pred = rand_cv.predict(X_test)
print(pred)
print(sorted(rand_cv.cv_results_.keys()))
print('최적 하이퍼 파라미터: ', rand_cv.best_params_)
print('최고 예측 정확도: {:.4f}'.format(rand_cv.best_score_))
```

## 베이지안 옵티마이저
<ul>
<li>Bayesian Optimization는 베이즈 정리에 의하여 사후 확률을 통해 최적값을 예측해 최적의 파라미터를 업데이트</li>
<li>Bayesian Optimization 은 어느 입력값(x)를 받는 미지의 목적 함수(f(x))를 상정하여, 해당 함숫값(f(x))을 최대로 만드는 최적해를 찾음</li>
<li>즉, 목적 함수(탐색대상함수)와 하이퍼파라미터 쌍(pair)을 대상으로 Surrogate Model(대체 모델) 을 만들고, 순차적으로 하이퍼 파라미터를 업데이트해 가면서 평가를 통해 최적의 하이퍼파라미터 조합을 탐색</li>
<li>이 때의 목점 함수를 black-box function 이라고 한다.</li>
<br/>
<li>Bayesian Optimization은 두 가지 필수 요소가 존재</li>
<li>1) Surrogate Model은, 현재까지 조사된 입력값-함숫결과값 점들 (x1, f(x1)),...,(xt, f(xt)) 을 바탕으로, 미지의 목적 함수의 형태에 대한 확률적인 추정을 수행(사전확율)</li>
<li>2) Acquisition Function은, 목적 함수에 대한 현재까지의 확률적 추정 결과를 바탕으로, '최적 입력값을 찾는 데 있어 가장 유용할 만한' 다음 입력값 후보를 추천(사후확률로 업데이트) </li>
</ul>
