# 17. 데이터 프레임 합치기

```python
air_quality_no2 = pd.read_csv("https://raw.githubusercontent.com/haram4th/ablearn/main/air_quality_no2_long.csv", parse_dates=True)
air_quality_no2 = air_quality_no2[["date.utc", "location", "parameter", "value"]]
air_quality_no2.head()
```

```python
air_quality_pm25 = pd.read_csv("https://raw.githubusercontent.com/haram4th/ablearn/main/air_quality_pm25_long.csv", parse_dates=True)
air_quality_pm25 = air_quality_pm25[["date.utc", "location", "parameter", "value"]]
air_quality_pm25.head()
```

![08_concat_row](./images/08_concat_row.png)

## 1) concat: 데이터 프레임 전체를 합치기

```python
air_quality = pd.concat([air_quality_pm25, air_quality_no2], axis=0)
air_quality.head()
```

```python
print('Shape of the ``air_quality_pm25`` table: ', air_quality_pm25.shape)
print('Shape of the ``air_quality_no2`` table: ', air_quality_no2.shape)
print('Shape of the resulting ``air_quality`` table: ', air_quality.shape)
```

```python
air_quality = air_quality.sort_values("date.utc")
air_quality.head()
```

```python
air_quality_ = pd.concat([air_quality_pm25, air_quality_no2], keys=["PM25", "NO2"])
air_quality_.head()
```

## 2) Join: key 컬럼을 중심으로 2개의 데이터 프레임을 합침

![08_merge_left](./images/08_merge_left.png)

```python
stations_coord = pd.read_csv("https://raw.githubusercontent.com/haram4th/ablearn/main/air_quality_stations.csv")
stations_coord.head()
```

```python
air_quality.head()
```

```python
air_quality = pd.merge(air_quality, stations_coord, how="left", on="location")
air_quality.head()
```

```python
air_quality_parameters = pd.read_csv("https://raw.githubusercontent.com/haram4th/ablearn/main/air_quality_parameters.csv")
air_quality_parameters.head()
```

```python
air_quality = pd.merge(air_quality, air_quality_parameters, how='left', left_on='parameter', right_on='id')
air_quality.head()
```
