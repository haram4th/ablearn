텍스트 정리, 불필요한 기호 정리하기

```python
import string
```

```python
string.punctuation
```

```python
data['text'][2967245]
```

```python
# 특수문자 제거 함수 만들기
def clean(x):
  cleaned = []
  for i in x:
    if i not in string.punctuation:
      cleaned.append(i)
  cleaned = "".join(cleaned)
  return cleaned
```

```python
data['text'] = data['text'].apply(clean)
```

```python
data
```

```python
data2 = data.copy()
```

stop words 상투적인 단어들 제거 

```python
from nltk.corpus import stopwords
```

```python
import nltk
nltk.download('stopwords')
```

```python
# stopword 제거하는 함수
def stopwd(x):
  words = []
  for i in x.split():
    if i.lower() not in stopwords.words('english'):
      words.append(i.lower().replace('place', "").replace('ive', "").replace('im', ""))
  return words
```

```python
data['text'] = data['text'].apply(stopwd)
```

나이브 베이즈로 사용자 리뷰 텍스트 감성분석하기

```python
data2 = data.copy()
```

```python
data2.head()
```

```python
X = data2['text']
y = data2['stars']
```

```python
X
```

```python
y
```

단어를 CountVectorizer를 사용해서 one-hot encoding 해주기

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer()
cv.fit(X)
```

```python
X = cv.transform(X)
```

```python
X
```

훈련 테스트 데이터로 분리

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.2, random_state=77)
```

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
model = MultinomialNB()
```

```python
model.fit(X_train, y_train)
```

```python
pred = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
print(accuracy_score(y_test, pred))
```

```python
print(confusion_matrix(y_test, pred))
```

```python
print(classification_report(y_test, pred))
```

랜덤포레스트로 분석해보기 (비교를 위해서)

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier(max_depth = 10, n_estimators= 1000, random_state= 77, n_jobs= -1 )
```

```python
rf.fit(X_train, y_train)
pred2 = rf.predict(X_test)
```

```python
print(accuracy_score(y_test, pred2))
print(confusion_matrix(y_test, pred2))
print(classification_report(y_test, pred2))
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
params = { 'max_depth' : [5, 10, 15], 'n_estimators' : [100, 200, 300] }
```

```python
rand_cv = RandomizedSearchCV(rf, param_distributions= params, cv=3, n_jobs=-1, random_state= 77)
rand_cv.fit(X_train, y_train)
pred = rand_cv.predict(X_test)
print(sorted(rand_cv.cv_results_.keys()))
print('최적 하이퍼 파라미터: ', rand_cv.best_params_)
print('최고 예측 정확도: {:.4f}'.format(rand_cv.best_score_))
```
