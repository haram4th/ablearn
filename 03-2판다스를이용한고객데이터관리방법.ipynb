## Description 컬럼의 데이터 정제 필요
<li>문자형이 아닌 값을 문자로 변환</li>
<li>특수문자, 2칸이상 공백 제거</li>
<li>소문자로 변환</li>

```python
description = data["Description"].unique()
```

```python
for index, cont in enumerate(description):
  if len(str(cont)) < 8:
    print(index, cont)
```

```python
# Description 컬럼의 데이터를 모두 문자열로 변환
data["Description"] = data["Description"].apply(lambda x : str(x))
```

```python
# string 모듈을 임포트 해서 특수문자 삭제
import string
data["Description"] = data["Description"].apply(lambda x: x.translate(str.maketrans("","",string.punctuation)))
```

```python
# Description 컬럼 내용 양 옆의 공백 삭제
data["Description"] = data["Description"].str.strip()
```

```python
# nan 값을 None 으로 교체
data["Description"] = data["Description"].str.replace("nan", "None")
```

```python
# 대문자를 모두 소문자로 변경
data["Description"] = data["Description"].str.lower()
```
